# Analisi vaccini
### Confronto Tra i vaccini Astrazeneca, Sputnik, Pfizer  in America ,Europa, Russia

```python
from bsi_sentiment.twitter import search_tweets_sn
import pandas as pd
```

#### Estrazione dati

L'estrazione avviene tramite script scraping. Lo script prevede come input:

1) la parola da cercare (unsensitive)

2) la città di riferimento

3) il raggio in km del cerchio all'interno del quale cerchiamo i tweets 

4) il numero totale di tweet

5) Periodo Di Estrazione


### Abbiamo scelto 3 macro regioni

America: SaltLake City e Indianapolis con raggio 1100 km

Europa: francoforte 760 km + Are (città in finlandia) con raggio 800km

Russia : Moscow raggio 700 km

#### Funziona di Estrazione dati

```python
def extractDataAndSaveInCsv(citta, vaccino,raggio,location):
    # prelevo i tweets
    tweets = search_tweets_sn(
      q=vaccino,
      since="2021-03-01",
      until="2021-03-31",
      near=citta,
      radius=raggio,
      lang="en",
      max_tweets=10
    )
    
    if(len(tweets.tweets)!=0):
        # salvo il dataframe
        directory = './dataset/dataset_tweets/'
        file= "tweet_"+"marzo"+vaccino+"_"+citta+"_"+raggio+".csv"
        tweets.to_csv(path= directory + file)

        #aggiungo la location
        df = pd.read_csv(directory+file)
        df['location'] = location
        return df
    return print("nessun tweets trovato con vaccino "+ vaccino)
```

#### Estrazione dati America

#### Indianapolis

```python
citta = "Indianapolis"
raggio="1100km"
location ='America'
```

```python
## Astrazeneca

vaccino="AstraZeneca"
extractDataAndSaveInCsv(citta,vaccino , raggio , location)

## Sputnik

vaccino="Sputnik"
extractDataAndSaveInCsv(citta,vaccino , raggio , location)

## Pfizer 
vaccino="Pfizer"
extractDataAndSaveInCsv(citta,vaccino , raggio , location)
```
