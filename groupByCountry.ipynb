# Aggiunta righe mancanti

```python
def aggiungeRigheDataMancanti(df,area,tag):
    idx = pd.date_range('03-01-2021', '03-31-2021')
    df.loc[:,'date'] = pd.to_datetime(df.loc[:,'date'])
    df = df.set_index(df.loc[:,'date'])

    df = df.reindex(idx, fill_value=0)

    df.loc[:,'area']=area
    df.loc[:,'parola_cercata']=tag
    df['date'] = df.index
    df.reset_index(level=0, inplace=True)
    df = df.drop(['index'], axis=1)
    df
    return df
```
